
# **API con Flask y modelo de ML o DL**

El objetivo de este proyecto es que cada grupo desarrolle una **API en Flask** que incorpore un **modelo de Machine Learning o Deep Learning**, y que finalmente se despliegue en un entorno accesible públicamente.  
El despliegue puede hacerse en **AWS EC2**, **Render** o **PythonAnywhere**, según lo que cada equipo prefiera.

La idea es que se construya un servicio sencillo pero funcional: una API que reciba datos, los procese con un modelo y devuelva una predicción. No se busca complejidad extrema, sino que seáis capaces de recorrer el ciclo completo: desarrollo, integración del modelo y despliegue.

---

# Requisitos del proyecto

### Backend Flask
- Crear una API REST con al menos tres rutas:
  - Una ruta de estado del servicio (`/health`)
  - Una ruta con parámetros en el **path**
  - Una ruta con parámetros en la **query**
  - Una ruta que reciba datos en el **cuerpo** (JSON)  
- Manejo básico de errores y respuestas claras

### Modelo de ML o DL
- Puede ser un modelo entrenado por vosotros o uno preentrenado (por ejemplo, de `sklearn` o un modelo ligero de Keras)
- Debe cargarse al iniciar la aplicación y usarse dentro de un endpoint

### Despliegue
- Elegir una de estas plataformas:
  - **AWS EC2**  
  - **Render**  
  - **PythonAnywhere**  
- La API debe quedar accesible mediante una URL pública

### Entregables
- Nombres y apellidos de los integrantes del equipo
- Repositorio con el código
- Archivo `requirements.txt`
- Instrucciones para ejecutar el proyecto en local
- URL del servicio desplegado
- Ejemplos de peticiones y respuestas
- Breve documentación técnica
