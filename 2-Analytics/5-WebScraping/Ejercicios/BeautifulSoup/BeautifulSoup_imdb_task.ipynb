## Web scraping de IMDB

Descarga la información correspondiente y guarda en un csv el top de las 250 películas mediante webscraping. Encapsúlalo en un script.

Obtén:
* Título
* Año
* Duración
* Posición
* Rating

```python
import cloudscraper     # de momento funciona para que Cloudfare no nos bloquee
from bs4 import BeautifulSoup

scraper = cloudscraper.create_scraper(
    browser={
        'browser': 'chrome',
        'platform': 'android',
        'desktop': False
    }
)
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}

html = scraper.get("https://www.filmaffinity.com/es/countcat.php?id=new_hbo_es")

soup = BeautifulSoup(html.text, "html.parser")
print(soup.title)
```

```python
soup = BeautifulSoup(html.text, "html.parser")
# top 25
soup
```

```python
duracion = soup.find("div", class_='duration').get_text(strip=True)
duracion
```

```python
#TITULO
for i in soup.find_all("a", class_='d-none d-md-inline-block'):
    print(i.text)
```

```python
#DURACION
for i in soup.find_all("div", class_='duration'):
    print(i.get_text(strip=True))
```

```python
#RANKING
for i in soup.find_all("div", class_='fa-list-position'):
    print(i.get_text(strip=True))
```

```python
#RATING
for i in soup.find_all("div", class_='avg mx-0'):
    print(i.get_text(strip=True))
```

```python
dict = {
    "titulo" : [],
    "año" : [],
    "duracion" : [],
    "posicion" : [],
    "rating" : []
}
```

```python
# TODO  

#TITULO
for i in soup.find_all("a", class_='d-none d-md-inline-block'):
    titulo = i.text
    dict["titulo"].append(titulo)

#DURACION
for i in soup.find_all("div", class_='duration'):
    duracion = i.get_text(strip=True)
    dict["duracion"].append(duracion)

#RANKING
for i in soup.find_all("div", class_='fa-list-position'):
    posicion = i.get_text(strip=True)
    dict["posicion"].append(posicion)

#AÑO
for i in soup.find_all("span", class_='mc-year'):
    año = i.get_text(strip=True)
    dict["año"].append(año)

#RATING
for i in soup.find_all("div", class_='avg mx-0'):
    rating = i.get_text(strip=True)
    dict["rating"].append(rating)
```

```python
dict
print(len(dict["titulo"]))
print(len(dict["año"]))
print(len(dict["posicion"]))
print(len(dict["duracion"]))
print(len(dict["rating"]))
```

```python
import pandas as pd
df = pd.DataFrame(dict)
df
```

```python
# código

df.to_csv("top25.csv")
df
```
