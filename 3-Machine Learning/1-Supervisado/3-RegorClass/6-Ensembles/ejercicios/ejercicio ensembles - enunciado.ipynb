### 3. Bagging
Para este apartado tendrás que crear un ensemble utilizando la técnica de bagging ([BaggingClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.BaggingClassifier.html)), mediante la cual combinarás 100 [DecisionTreeClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html). Recuerda utilizar también [cross validation](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html) con 10 kfolds.

**Para este apartado y siguientes, no hace falta que dividas en train/test**, por hacerlo más sencillo. Simplemente divide tus datos en features y target.

Establece una semilla

````python
semilla = 42

array = df.values

X = array[:, 0:8]

y = array[:, 8]
````

````python
# importa el árbol de clasificación, un clasificador de tipo bagging y un kfold
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

````

````python
dtc = DecisionTreeClassifier()
estimadores = 100
model = BaggingClassifier(estimator=dtc, n_estimators=estimadores, random_state=semilla)

# haz un árbol de clasificación con profundidad 5
# haz un bagging classifier con ese tipo de árbol y 100 estimadores
````

````python

kfold = KFold(n_splits=10, shuffle=True, random_state=semilla)

# no te tiene por qué salir igual
````

````python
results = cross_val_score(model, X, y, cv=kfold, scoring='accuracy')

mean = results.mean()
std = results.std()

print(mean)
print(std)
````

### 4. Random Forest
En este caso entrena un [RandomForestClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html) con 100 árboles y un `max_features` de 3. También con validación cruzada

````python
from sklearn.ensemble import RandomForestClassifier
````

````python
estimadores = 100
max_features = 3
model_rf = RandomForestClassifier(n_estimators=estimadores, max_features=max_features, random_state=semilla)

results_rf = cross_val_score(model_rf, X, y, cv=kfold, scoring='accuracy')

mean_rf = results_rf.mean()
std_rf = results_rf.std()

print(mean_rf)
print(std_rf)
````

### 5. AdaBoost
Implementa un [AdaBoostClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.AdaBoostClassifier.html) con 30 árboles.

````python
from sklearn.ensemble import AdaBoostClassifier
````

````python
estimadores = 30

model_ada = AdaBoostClassifier(n_estimators=estimadores, random_state=semilla)
results_ada = cross_val_score(model_ada, X, y, cv=kfold, scoring='accuracy')

mean_ada = results_ada.mean()
std_ada = results_ada.std()

print(mean_ada)
print(std_ada)
````

### 6. GradientBoosting
Implementa un [GradientBoostingClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.GradientBoostingClassifier.html) con 100 estimadores

````python
from sklearn.ensemble import GradientBoostingClassifier
````

````python
estimadores = 100

model_gbc = GradientBoostingClassifier(n_estimators=estimadores, random_state=semilla)
results_gbc = cross_val_score(model_gbc, X, y, cv=kfold, scoring='accuracy')

mean_gbc = results_gbc.mean()
std_gbc = results_gbc.std()

print(mean_gbc)
print(std_gbc)
````

### 7. XGBoost
