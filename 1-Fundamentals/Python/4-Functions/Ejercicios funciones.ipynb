Escribe una función que convierta números del 1 al 7 en nombres de los dias de la semana. La función constará de un único argumento numérico y una salida de tipo string

## Ejercicio 3
Escibe una función que compare dos números. La función tiene dos argumentos y hay tres salidas posibles: que sean iguales, que el primero se  mayor que el segundo, o que el segundo sea mayor que el primero

````python

def comparador(i,j):
    if i == j:
        return f"{i} es mayor que {j}"
    if i > j:
        return f"{i} es mayor que {j}"
    if i < j:
        return f"{i} es menos que {j}"
    else:
        return
    
comparador(10,8)
    
    
````

## Ejercicio 5
Escribe una función que tenga un único argumento, un string. La salida de la función tiene que ser un diccionario con el conteo de todas las letras de ese string.

````python
def cadena(palabra):
    total = {}
    for i in palabra:
        if i in total:
            total[i] += 1
        else:
            total[i] = 1
    return total

cadena("todas las letras")
    
````

## Ejercicio 8
Escribe un programa que obtenga el enésimo número de la [serie de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci). Tienes que crear una función recursiva con un único argumento.

```Python
fibonacci(2) = 1
fibonacci(4) = 3
fibonacci(5) = 5
fibonacci(8) = 21
```

````python
# calcula el numero fibo

def fibonacci(numero):
    if numero == 0: 
        return 0
    if numero == 1:
        return 1
    return fibonacci(numero-1) + fibonacci(numero-2)

# calcula hasta llegar al fibo

def serie_fibonacci(n):
    lista = []
    for i in range(n+1):
        lista.append(fibonacci(i))
    return lista


fibonacci(3)
serie_fibonacci(30)
````

Define en una única celda las siguientes funciones:
* Función que calcule el área de un cuadrado
* Función que calcule el area de un triángulo
* Función que calcule el área de un círculo

En otra celda, calcular el area de:
* Dos círculos de radio 10 + un triángulo de base 3 y altura 7
* Un cuadrado de lado = 10 + 3 círculos (uno de radio = 4 y los otros dos de radio = 6) + 5 triángulos de base = 2 + altura = 4

````python
#1 Ejercicio 1

from math import pi

def area_cuadrado(i):
    return i **2

def area_triangulo(i,j):
    return (i * j) / 2

def area_circulo(i):
    return pi * i ** 2

print(area_cuadrado(3))
print(area_triangulo(3,4))
print(area_circulo(3))
````
