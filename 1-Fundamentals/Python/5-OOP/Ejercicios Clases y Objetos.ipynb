Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int

* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada.
```python
class Tienda:
    '''
    Creamos atributos genericos de tipo y si la tienda esta abierta.
    '''

    tipo = "Electrodomesticos"
    abierta = True

    def __init__(self, nombre, direccion, numero_empleados, ventas_3meses):

        self.nombre = nombre
        self.direccion = direccion
        self.numero_empleados = numero_empleados
        self.ventas_3meses = ventas_3meses
    
    def ventas_todos_meses(self):
        '''
        Metodo para sumar ventas de los últimos 3 meses.
        '''
        return sum(self.ventas_3meses)
    
    def media_3ventas_empleados(self):
        '''
        Metodo para calcular la media de venta por empleado últimos 3 meses.
        '''
        return self.ventas_todos_meses() / self.numero_empleados
    
    def nombre_completo(self):
        '''
        Metodo devuelvo el nombre completo y direccion.
        '''
        return f"{self.nombre} {self.direccion}"

    def venta_ultimo_mes(self):
        '''
        Metodo devuelve la venta del ultimo mes.
        '''
        return self.ventas_3meses[-1]
    
    def proyeccion(self, inversion):
        '''
        Metodo que devuelve la proyeccion de ventas en base a la inversion en marketing.
        '''
        cambio = []
        for i in self.ventas_3meses:
            if inversion > 1000:
                cambio.append(i*1.2)
            else:
                cambio.append(i*1.5)
        self.ventas_3meses = cambio
        return cambio  

t1 = Tienda("MediaMarkt Alcorcon","CC. Parque Oeste, Av. de Europa,22,28922 Alcorcón,Madrid", 88, [ 2.3 , 3.5 , 4])
t2 = Tienda("MediaMarkt Fan Mallorca","CC. Fan Mallorca Avenida Alcover ,37,07007 Palma de Mallorca", 110, [ 3.8 , 4.1 , 4.3])
t3 = Tienda("MediaMarkt Francesc Macià","Avenida Diaognal ,477,08021 Barcelona", 23, [ 1.1 , 1.5 , 1.8])

print(t3.tipo)
print(t3.abierta)
print(t3.media_3ventas_empleados())
print(t3.nombre_completo())
print(t3.venta_ultimo_mes())
print(t3.proyeccion(1400))

todas_tiendas = [t1 , t2 , t3]
total_ventas = 0
for i in todas_tiendas:
    total_ventas = total_ventas + i.venta_ultimo_mes()


for i in todas_tiendas:
    if "Avenida" in i.direccion:
        print(i.nombre)

```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
class Perro:
    '''
    Valores por defecto que un perro tiene. Atributos compartidos.
    '''
    patas = 4
    orejas = 2
    ojos = 2
    velocidad = 0

    def __init__(self, nombre, raza, pelo="Marron", dueño = None):
        '''
        Atributos propios de cada 1 de los perros.
        '''
        self.nombre = nombre
        self.raza = raza
        self.pelo = pelo
        self.dueño = dueño

    def andar(self,aumento_velocidad):
        '''
        Metodo para que el perro comience a caminar(andar)
        '''
        self.velocidad = self.velocidad + aumento_velocidad
        return self.velocidad

    def parar(self):
        '''
        Metodo que lo para
        '''
        self.velocidad = 0
        return self.velocidad

    def ladrar(self, palabra):
        '''
        Metodo que devuelve un ladrido y la frase que escribamos en la funcion.
        '''
        return f"GUAU! {palabra}"
    
perro1 = Perro("Pongui","Beagle","Tricolor","Beatriz")

print(perro1.nombre)
print(perro1.patas)
print(perro1.raza)
print(perro1.pelo)
print(perro1.dueño)

print(perro1.ladrar("tengo hambre"))
print(perro1.andar(5))
print(perro1.andar(10))
print(perro1.parar())
print(perro1.andar(10))
print(perro1.andar(10))

```
